# Rules for which brackets/parenthesis to use
* If it's a function you're calling, you need parenthesis.  For example, print() is a function. So it needs parenthesis.
* If you are referencing  a list you need square brackets, e.g. howmuch[0]
* If you are creating a list you also need square brackets, e.g. fruits = ["bananas", "apples", "pears"]

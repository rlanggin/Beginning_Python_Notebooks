````python
x=13
while x < 17:
    print (x)
    x=x+1
````

##  Function to convert temperature from Fahrenheit to celsius

````python
# This is code that does the conversion, but is not a function.
tempfah = 32
tempcel = (tempfah - 32)*(5/9)
print ("The temperature in celsius is", tempcel)
````

````python
# Let's define a function that does the conversion for us.
def convtemp2celsius(f):    # f is the argument, it's the placeholder for the temperature
                            # in fahrenheit
    tempcel = (f - 32)*(5/9)   # here's the calculation that converts fahrenheit to celsiu
    print(tempcel)        # here's the line that returns the celsius temperature
````

````python
convtemp2celsius(212)
````

````python
print(convtemp2celsius(212))
````

````python
print(convtemp2celsius(32))
````

````python
print(convtemp2celsius(80))
````

````python
print(convtemp2celsius(-40))
````

So we created a function called "convtemp2celsius" whose job is to convert fahrenheit temperatures to celsius temperatures.  We can _pass_ the function any number we want, and it will convert that number to celsius.   You can see how this would be useful if we find ourselves converting a lot of temperatures.

# Difference between "return" and "print" in a function
This is a semi-annoying difference. It's sort of subtle.  You can actually see the difference in the two cells above "Lingo". ```range(7)``` actually produced an ``Out[]`` line, whereas ``print("20")`` didn't.

### Difference #1
```return()``` produces **Output**, while ```print()``` doesn't.

### Difference #2
```print()``` will always print to the screen while ```return()``` will only print to the screen sometimes.

Watch this:

````python
range(7)
print("20")
````

Notice that ```range(7)``` didn't produce anything on the screen, while the print statement did. ```range(7)``` will only print something to the screen if it's the last thing in the cell.  Like this..

````python
print("20")
range(7)
````

Notice that the ```print()``` statement still printed something to the screen even though it wasn't last.

## Difference #3
If a function has a ```return()``` statement in it, you can set the output of the function equal to something.
Check out the following function that squares a number:

````python
"""Define a function that squares a number"""
def sqr(a):
    return (a**2)
````

````python
"""Call that function 4 times, making the output of one call into the input of the next"""
x=2
x = sqr(x)  # Square the number and replace startingnum with that new number
x = sqr(x)
x = sqr(x)
x = sqr(x)
print (x)
````

````python
"""Define a function that squares a number and just prints it without returning it"""
def sqr_wo(a):
    print (a**2)
````

````python
x=2
x=sqr_wo(x)
x=sqr_wo(x)
x=sqr_wo(x)
x=sqr_wo(x)
print(x)
````

## Notice it didn't even like calling ```sqr_wo(x)``` after I set ```x=sqr_wo(x)``` because at that point x didn't equal anything, because there was no official output of the function.

## A word about function definitions.
The function definition always includes the name of a variable (called a parameter), such as ```a``` in the cases above.  When I call the function I can use any variable I want.  Defining a function is sort of like saying "whatever number I give you, this is what I want you to do with it.")

````python
## Bonus assignment - add a return statement
def sqr(x):   # defining a function that squares a number
    s = x**2
    return(s)  # Now it returns the square instead of printing it
````

Since I returned the value instead of just printing it, I have some more options for using it.
* I can use it in a subsequent calculation.
* I can use the function in a print statement.
* I can create a list out of the things it returns.

````python
# Use it in a subsequent calculation
sqr(sqr(10))
````

````python
# Use it in a print statement
print("The square of 10 is", sqr(10))
````

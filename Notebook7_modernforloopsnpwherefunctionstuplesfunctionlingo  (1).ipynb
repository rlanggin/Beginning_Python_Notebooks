# My assignment to myself
Define a function that takes two arguments, a string and a number.
The function should return a tuple that consists of three items, the string, the number, the number of characters in the string, and the product of the number and the number of characters in the string.
Pass "hello" and "4" to the function.


```python
def weird(x,y):  # define a function weird that returns the same
    # string and number you gave it, and also the number of characters
    # in that string, and the produce of the number you gave it times
    # the number of characters in the string
    mytuple=(x,y,len(x),y*len(x))
    return(mytuple)
```

```python
# First let's try just calling weird
weird("haverford", 10) # Calling the function weird
```

Notice it returns a tuple. (You can tell because it has parentheses around it.)

If you only want one piece of the tuple there are several ways to go about it.

### Method 1

```python
weirdout = weird("hello",4)  #Setting the output of weird("hello", 4)
# equal to weirdout so that I can use weirdout for things
weirdout
```

So now weirdout is set to be that tuple.

```python
# So if we just want the 3rd item in the tuple we say
weirdout[3]
```

### Method 2
You can accomplish the same thing all in one line by saying....

```python
weird("hello",4)[3]
```

So that "[3]" at the end of the call to weird, tells it we only want the 3rd item in the tuple that it returns.

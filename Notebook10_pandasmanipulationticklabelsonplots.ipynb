###  Finally, you could figure out a way to create a daynumber instead of this date "string" which is kind of annoying.

````python
# Look at one date to remind ourselves of the format
shortdf.date[0]
````

````python
# the "expand=True" tells it to create a new colum for each
# the pat = "-" tells it what character to use the split the strong
shortdf.date.str.split(pat = "-", expand=True)
````

````python
shortdf.date.str.split(pat = "-", expand=True)[2].astype(int)
````

We've got it split into columns now, but we can't do math with them, because they're strings. (It would be like asking python to multiply 3 by spaghetti.). So we're going to "recast" them as integers.

The "[1]" means 1st column and the "astype(int)" converts it from a string into an int.

````python
shortdf.date.str.split(pat = "-", expand=True)[1].astype(int)
````

````python
# So to create a "daynumber" you could multiply the month by the number of days
# in a month and then add the day.
# The "\" character is used for telling python to continue on the next line.
daynumber = shortdf.date.str.split(pat = "-", expand=True)[1].astype(int)*30 +   \
     shortdf.date.str.split(pat = "-", expand=True)[2].astype(int)
````

I wanted to make a function that would let me plot the data for any country.  It follows:

````python
# Okay and now we're ready to make a function that plots using this
def plotdf_daynumber(df, country, keyword):
    shortdf=df[df['location'].str.contains(country)]
    daynumber=[]
    for i in shortdf.index:
        month = int(shortdf.loc[i].date.split("-")[1]) # Get the month number
        if (month == 1 or month==3 or month==5 or month==7 or month==9 or month==1): # get all the months that have 31 days
            daysinmonth=31
        elif month==2 :
            daysinmonth=28
        else :
            daysinmonth=30
        daynumber.append(int(shortdf.loc[i].date.split("-")[0])*365 +
        int(shortdf.loc[i].date.split("-")[1])*daysinmonth +
         int(shortdf.loc[i].date.split("-")[2]))

    daynumber = np.array(daynumber) # The next line doesn't work if I don't do this
    daynumber = daynumber - min(daynumber)
    plt.plot(daynumber,
         df[df['location'].str.contains(country)][keyword]);
    plt.xlabel("daynumber since start of pandemic")
    plt.ylabel(keyword)
    plt.title(country)
````

````python
# Let's test it:
plotdf_daynumber(Covid, "United States", "new_cases")
````

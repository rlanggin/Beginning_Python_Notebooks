## Let's use an example from the last notebook to motivate this idea of a loop...

````python
howmuch=[5, 7, 9]
fruit = ["apples", "pears", "bananas"]
print("Grocery list")
print(howmuch[0], fruit[0])
print(howmuch[1], fruit[1])
print(howmuch[2], fruit[2])
````

And then I encouraged you to get lazier:

````python
i=0
print("Grocery list")
print(howmuch[i], fruit[i])
i=1
print(howmuch[i], fruit[i])
i=2
print(howmuch[i], fruit[i])
````

Which produced the same output!

So let's get lazier. Remember the principle is to make the computer do the work for you.
We want the computer to count from 0 to 2 for us, and print ```howmuch[i]```, ```fruit[i]``` as i goes from 0 to 2.
So in other words, for i in the range 0 to 2, we want to print ```howmuch[i]```, ```fruit[i]```.

Here's the syntax:

````python
for N in range(0,3):
    print(howmuch[N], fruits[N])
````

We played around with this for awhile. I wanted to show you that if you include a third number in the range statement, that it is the stepsize.  See below.

````python
for N in range(0,200,13):  # First number is starting, second number is ending
    # (but not including) and third number is the step size
    print("This time through the loop N = ", N)
    print("hello Rachel")
````

````python
print(N)  # We were a little surprised that this printed 195, but it makes sense,
# because that's the last thing that "N" was assigned to.
````

````python
# The following was in response to a question Joey asked about a colon between numbers instead of a comma.  You need a colon
# when it's a range of references to a list.  But the comma is what the "range" function wants.  That's just kind
# of one of those annoying things you just need to know.
fruits[0:2]
````

In the above we learned:
    * a) that range(0,2) only gives you 0,1
    * b) that range(2) is a shorthand for range(0,2)
    * c) that range(0,10,2) gives you 0,2,4,6,8

We sort of went off the rails here, but it was fun.  I wanted to show you two different ways to write loops using musicals and their characters.

````python
musicals=["Six", "Music Man", "Wicked", "Phantom of the Opera", "Waitress", "Hades Town", "Kinky Boots"]
````

````python
musicals
````

````python
# Here's how you can create a counter that counts the number of entries in a list.
counter=0
for musical in musicals:
    print(musical)
    counter+=1 # Same thing as counter=counter+1
    print(counter)
````

````python
characters=["Jane Seymour", "Professor Hill", "Elphaba", "Christine", "Shirley", "Satan", "Charlie"]
````

I wanted to show you two different ways to write which characters were in which musicals.  The first way uses the "python-y" style of loop.  The second way uses the numeric indexing.  There are undoubtedly more ways to do this. Definitely write me if you find a way you like that I didn't mention!

````python
i=0 # You start off the counter with i=0
for character in characters:
    print(character, "is in the musical", musicals[i])
    i+=1 # Each time you go through the loop you increase the counter by 1
````

````python
# This produces the same output using fewer lines.  That doesn't mean it's better necessarily,
# but knowing both ways will make you a more flexible programmer.
for i in range(7):
    print(characters[i], "is in the musical", musicals[i])
````

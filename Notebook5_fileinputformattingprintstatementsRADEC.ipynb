# Fancy print statements


The thing about fancy print statements is sometimes you want to put the values of several
variables into a print statement.  If you want to have control over how the values print
you need the following:

````python
# If you wanted to print "Hello 14" you can do it this way:
print("Hello 14!")
````

````python
# But what if you didn't know the value of x ahead of time and you wanted it to print x
# in place of the 14?
x = (37.0 + 42.0)*1.3
print("Hello %d!" %x)
````

The way this works is the ```%d``` inside the quoted expression tells Python something like "Save
this spot for something I'm about to tell you to print."

The "d" in particular tells Python that "x" is going to be an integer.
* d is for integer
* f is for "floating point" (something with a decimal)
* s is a string, aka a word

Let's try all of them:

````python
print("Hello %d!" %x)
print("Hello %f!" %x)
print("Hello %s!" %x)
````

Watch what happens when I try to set x equal to a word (aka a "string"):

````python
x= "there"
print("Hello %d!" %x)
````

But this will work, because "there" is a "string" (aka a string of characters.)

````python
print("Hello %s!" %x)
````

What else can I specify around the f?

The ```f``` format in particular lets you specify how many digits should be after the decimal point.

````python
x = 37.0
print("Hello %.0f!" %x)
print("Hello %.1f!" %x)
print("Hello %.2f!" %x)
print("Hello %.3f!" %x)
````

You can also put a number before the decimal point to tell Python how big a field you want it to leave for that number.

````python
print("Hello %4.1f!" %x)
print("Hello %5.3f!" %x)
print("Hello %6.1f!" %x)
print("Hello %7.1f!" %x)
````

Finally, if you would like to specify multiple numbers you can! You have to put parenthesis around them to indicate it's a "tuple". (We haven't talked about tuples.  It's tuple like quadruple, or quintuple....it's a bunch of things squished together.)

````python
x = 14.3
y = 2
print("I need %14.1f apples and %d pears and %s bananas." %(x,y,"a ton of"))
````

````python
"""Define a function that accepts age as an argument and returns the year you were born. The
function takes three arguments: your age, your birthmonth, and the current month.  Birthmonth and
curmonth must be integers (not strings - challenge assignment, make this function accept strings for months!).
It compares birthmonth to curmonth to determine if you've had your birthday yet this year.  Second challenge:
Modify the function so that it also take the day of the month."""
def yearborn(age, birthmonth, curmonth):
    yr = 2021-age    # subtract the age from 2020
    if (curmonth<birthmonth): # Decides if you've had your birthday yet this year
        yr -= 1    # new syntax!!!  That's short for yr = yr -1
    print("You were born in %d." %yr)
    return(yr)       # return yr
````
